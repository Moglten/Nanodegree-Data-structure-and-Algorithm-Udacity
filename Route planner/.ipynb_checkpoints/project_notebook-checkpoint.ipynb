````python
import heapq
from math import pow ,sqrt


def shortest_path(input_map, start, goal):
    if goal == start:
        return [start]

    # get all roads in 2D list
    roads = input_map.roads
    coordinates = input_map.intersections
    visited = {start}
    explored = [start]
    frontier_heap = []
    distance_from_goal = get_estimated_dist( coordinates, start, goal )
    current_city = (distance_from_goal, start)
    heapq.heappush( frontier_heap, current_city )

    while goal not in explored :

        if current_city[1] not in visited:
            current_city = heapq.heappop( frontier_heap )
            current_from_start = current_city[0] - get_estimated_dist( coordinates, current_city[1], goal )
            if current_city[1] in roads[explored[-1]]:
                explored.append( current_city[1] )

            visited.add(current_city[1])

            if current_city[1] == goal:
                return explored
            print(explored)
            # expand cities
            for city in roads[current_city[1]]:
                #h : distance from goal
                distance_from_goal = get_estimated_dist( coordinates, city, goal )
                #g : distance from start
                distance_path = path_dist( coordinates, city, current_city[1] ) + current_from_start
                #f : value that control
                city_value =  distance_from_goal + distance_path  
                
                heapq.heappush( frontier_heap, (city_value , city) )
       
    return explored


def get_estimated_dist(coordinate, city, goal):

    city_point = coordinate[city]
    goal_point = coordinate[goal]
    xs = city_point[0] - goal_point[0]
    ys = city_point[1] - goal_point[1]
    estimated_dist = sqrt(pow(xs,2) + pow(ys,2))

    return estimated_dist 


def path_dist(coordinate, intersection1, intersection2):

    inter1_point = coordinate[intersection1]
    inter2_point = coordinate[intersection2]
    xs = inter1_point[0] - inter2_point[0]
    ys = inter1_point[1] - inter2_point[1]
    path_dist = sqrt( pow( xs, 2 ) + pow( ys, 2 ) )

    return path_dist


shortest_path(map_40, 5, 24)



````
